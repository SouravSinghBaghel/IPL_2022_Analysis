 ## TEAM THAT WON MOST NUMBERS OF MATCHES WON

```python
sns.countplot(x="match_winner", data=df)
sns.set(rc={'figure.figsize':(20,10)})
plt.title("Number of Matches Won in IPL 2022")
plt.xlabel("Team")
plt.ylabel("Number of Matches Won")
plt.show()
```

## Number of number Won by using Defending or Chasing

```python
counts = df["won_by"].value_counts()
colors = ["red", "blue"]
fig, ax = plt.subplots(figsize=(6, 6))
plt.pie(counts, labels=counts.index, colors=colors, autopct="%1.1f%%", startangle=90)
plt.title("Number of number Won by using")
plt.show()





```

## Most Player of the Match Awards


```python
order = df["player_of_the_match"].value_counts().index
sns.catplot(x='player_of_the_match', kind='count', data=df, height=6, aspect=2.5)
plt.title('Most Player of the Match Awards')
plt.xticks(rotation=90)
plt.show()
```

## Top scorers of IPL 2022

```python
counts = df["top_scorer"].value_counts()
plt.bar(counts.index, counts)
plt.title("Top scorers of IPL 2022")
plt.xlabel("Player")
plt.xticks(rotation=90)
plt.ylabel("Count")
plt.show()
```

## Best Bowlers in IPL 2022

```python
order = df["best_bowling"].value_counts().index
sns.catplot(x='best_bowling', kind='count', data=df, height=6, aspect=2.5,order=order)
plt.title('Best Bowlers in IPL 2022')
plt.xticks(rotation=90)
plt.show()
```
